## Join matrices

```python
import pandas as pd
import os
import numpy as np
```

```python
path = '../data/'
matrix=pd.read_excel(os.path.join(path, 'Tn_library_log2FC_DASH.xlsx'))
matrix.head()
```

```python
matrix=pd.melt(matrix, id_vars=['Rv_ID', 'gene_name'], var_name='Expt', value_name='log2FC')
```

```python
matrix['Expt']=matrix['Expt'].str.replace('_log2FC', "")
```

```python
matrix=matrix[~matrix['log2FC'].isnull()]
```

```python
qvals=pd.read_excel(os.path.join(path, 'Tn_library_qvals_DASH.xlsx'))
qvals.head()
```

```python
qvals=pd.melt(qvals, id_vars=['Rv_ID', 'gene_name'], var_name='Expt', value_name='q-val')
qvals['Expt']=qvals['Expt'].str.replace('_q_val', "")
qvals.head()
```

```python
final=pd.merge(matrix,qvals, how='left', on=['Rv_ID', 'Expt', 'gene_name'])
```

```python
final[final['log2FC']==-np.inf]
```

```python
final.to_csv('Tn_library_DASH.csv', index=False)
```
