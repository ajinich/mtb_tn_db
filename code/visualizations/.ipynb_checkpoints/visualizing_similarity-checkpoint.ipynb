I want to visualize the binary vectors of Tn-library derived essentiality profiles (TnEP) on a 2-D "projection" where similar TnEPs are near each other. From a quick reading, what are some options to explore? 
* PCA - Principal component analysis
* tSNE - t-distributed Stochastic Neighbor Embedding
* non-negative matrix factorization (NMF)
* K-means clustering

What are the main steps I need to take? 
1. Play with demos of each of these, and choose the one that best fits what you want to do
2. Adapt the Tn-library data to the method of choice

```python
import numpy as np
import os
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
def get_cluster_coords(x1, x2, y1, y2, tn_tSNE):
    ind_cluster = [i for i in range(len(tn_tSNE)) if (tn_tSNE[i][0] > x1 and tn_tSNE[i][0] < x2 and tn_tSNE[i][1] > y1 and tn_tSNE[i][1] < y2 )]
    tn_cluster = tn_tSNE[ind_cluster]
    x_cluster = [X[0] for X in tn_cluster]
    y_cluster = [X[1] for X in tn_cluster]
    
    return ind_cluster, x_cluster, y_cluster

def essentiality_calls(df, gene_id):
# Return the columns in which a given gene is called essential:
    
    cols = df.columns[2:]
    col_list = []
    for col in cols:
        if df[df['Rv_ID']==gene_id][col].values[0] == 1:
            col_list.append(col)
            
    return col_list
```

## Trying to adapt tSNE functions to Tn-library dataset: 

```python
dir_data = '../../data'
```

```python
tn_file = os.path.join(dir_data, 'Tn_library_DB.xlsx')
df_tn = pd.read_excel(tn_file)
df_tn.shape
```

Merge with Sanger categories: 

```python
dir_annotations = os.path.join(dir_data, 'annotations')
df_sanger = pd.read_csv(os.path.join(dir_annotations, 'RvID_w_sanger_level_II.csv'))
df_sanger.shape
```

```python
df_tn = df_tn.merge(df_sanger, on = 'Rv_ID', how = 'left')
```

```python
data_cols = df_tn.columns[2:-1]
df_tn_data = df_tn[data_cols]
tn_array = df_tn_data.values
```

```python
tn_embedded = TSNE(n_components=2).fit_transform(tn_array)
```

```python
x = [X[0] for X in tn_embedded]
y = [X[1] for X in tn_embedded]
```

```python
df_tn['tSNE_1'] = x
df_tn['tSNE_2'] = y
```

```python
plt.figure(figsize = (10,10))
plt.plot(x, y, 'o', mec = 'k', ms = 7)
plt.tick_params(axis='both', which='both', bottom=False, top=False, left=False, labelleft=False, labelbottom='off')
plt.xlabel('t SNE - 1', fontsize = 24)
plt.ylabel('t SNE - 2', fontsize = 24)
```

Using "Hamming" distance: 

```python
tn_embedded_H = TSNE(n_components=2, metric='hamming').fit_transform(tn_array)
```

```python
xH = [X[0] for X in tn_embedded_H]
yH = [X[1] for X in tn_embedded_H]
```

```python
plt.figure(figsize = (20,15))
plt.plot(xH, yH, 'o', mec = 'k', ms = 7)
plt.tick_params(axis='both', which='both', bottom=False, top=False, left=False, labelleft=False, labelbottom=False)
plt.xlabel('t SNE - 1', fontsize = 24)
plt.ylabel('t SNE - 2', fontsize = 24)
```
