```python
import pandas as pd
import os
```

```python
lfc = pd.read_csv('../data/standardized_data/result_logfc_matrix_2019_12_10.csv')
```

```python
lfc.head()
```

```python
lfc=lfc.dropna(axis=0)
```

```python
lfc_vals=lfc.drop(columns=['Unnamed: 0'])
```

```python
from sklearn.decomposition import PCA
from plotnine import *
pca=PCA(3)
pca.fit_transform(lfc_vals)
```

```python
pca.explained_variance_
```

```python
pca_plot=pd.DataFrame(pca.fit_transform(lfc_vals))
```

```python
pca_plot.columns=['pca1', 'pca2', 'pca3']
```

```python
(ggplot(pca_plot, aes(x='pca1', y='pca2'))+geom_point())
```

```python
df=pca_plot.merge(lfc, left_index=True, right_index=True)
```

```python
df.sort_values(by='pca2', ascending=False)
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2).fit_transform(lfc_vals)
```

```python
tsne = pd.DataFrame(tnse, columns=['tsne1', 'tsne2'])
```

```python
(ggplot(tsne, aes(x='tsne1', y='tsne2'))+geom_point())
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(lfc_vals, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering 
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage ='ward')
y_hc=hc.fit_predict(lfc_vals)
```

```python
y_hc
```

```python
df_hc=lfc.copy(deep=True)
df_hc['hc'] = y_hc
```
